```python
import pandas as pd
import numpy as np
```

```python
columns = ['unit_number' , 'time', 'op_settings1' , 'op_settings2' , 'op_settings3'] + [f'sensor{i}' for i in range(1,22)]

data = pd.read_csv(f"C:\\Users\\acer\\Desktop\\Vangaurd\\data\\raw\\files\\train_FD001.txt", sep = r"\s+", header = None, names = columns) 

data
```

Calculating raw RUL and piece-wise  RUL for each engine

```python
data['max_cycle'] = data.groupby('unit_number')['time'].transform('max') 

data
```

```python
data['raw_RUL'] = data['max_cycle'] - data['time']

data['RUL'] = data['raw_RUL'].clip(upper = 125)

data
```

```python
data = data.drop(columns = ['max_cycle'])

data
```

Choosing sensors that actually correlates with the degradation of machinery

```python
sensors = [f'sensor{i}' for i in range(1,22)]
for sensor in sensors:
    deviation = data[sensor].std()
    print(f'{sensor} - {deviation}')
    
```

```python
#Sensors with zero or negligible standard deviation.
sensors_to_remove = ['sensor1', 'sensor5', 'sensor6', 'sensor10', 'sensor16', 'sensor18', 'sensor19']

data = data.drop(columns = sensors_to_remove)

data
```

```python
print(data.columns)
```

Calculating rolling window averages and standard deviation

```python
imp_sensors = ['sensor2', 'sensor3', 'sensor4', 'sensor7', 'sensor8', 'sensor9', 'sensor11', 'sensor12', 'sensor13', 'sensor14', 'sensor15', 'sensor17', 'sensor20', 'sensor21']

for sensor in imp_sensors:
    mean = data.groupby('unit_number')[sensor].transform( lambda x: x.rolling(window = 20, min_periods = 1).mean() )
    data[f'{sensor}_mean'] = mean

    std = data.groupby('unit_number')[sensor].transform( lambda x: x.rolling(window = 20, min_periods = 1).std() )
    data[f'{sensor}_std'] = std

data.head()
```

```python
data = data.fillna(0)
data.head()
```

Feature scaling and fitting feature scaler

```python
ex_cols = ['unit_number', 'time', 'RUL']

X = data.drop(columns=ex_cols)

Y = data['RUL']

print(X.columns.tolist())
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
x_array = scaler.fit_transform(X)

scaled_x = pd.DataFrame(x_array, columns = X.columns, index = X.index)

scaled_x
```

```python
import joblib
from pathlib import Path

models_path = f"C:\\Users\\acer\\Desktop\\Vangaurd\\models"
joblib.dump(scaler, models_path + '\\scaler.pkl')
```

```python
processed = Path("C:/Users/acer/Desktop/Vangaurd/data/processed")

processed_data = scaled_x.copy()
processed_data['RUL'] = Y

output_path = processed / 'processed_train_FD001.csv'
processed_data.to_csv(output_path, index = False)

processed_data
```
