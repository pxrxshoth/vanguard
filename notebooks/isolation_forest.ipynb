Experimentation of models for anomaly detection

```python
import mlflow
import pandas as pd
import joblib
from sklearn.ensemble import IsolationForest
from pathlib import Path
```

```python
base_path = Path("C:/Users/acer/Desktop/Vangaurd")

data = pd.read_csv(base_path / "data/processed/processed_train_FD001.csv")

print(data.shape)
print(data.columns.tolist)
```

```python
print(data.head())
print(data.columns.tolist())
```

```python
columns = ['unit_number' , 'time', 'op_settings1' , 'op_settings2' , 'op_settings3'] + [f'sensor{i}' for i in range(1,22)]

original_data = pd.read_csv(f"C:\\Users\\acer\\Desktop\\Vangaurd\\data\\raw\\files\\train_FD001.txt", sep = r"\s+", header = None, names = columns) 

original_data
```

```python
data['unit_number'] = original_data['unit_number']
data['time'] = original_data['time']

data
```

```python
x = data[data['time'] <= 30].copy()

print(x.columns.tolist())
```

```python
x
```

```python
ex_cols = ['unit_number' , 'time' , 'raw_RUL' , 'RUL']
x = x.drop(columns = ex_cols)
x
```

```python
print(x.columns.tolist())
```

```python
x = x.reset_index(drop = True)
```

```python

base_path = Path("C:/Users/acer/Desktop/Vangaurd")

mlflow.set_experiment("Vanguard_Anomaly_Detection")

with mlflow.start_run(run_name="Isolation Forest"):
    
    mlflow.log_param("model_type", "IsolationForest")
    mlflow.log_param("n_estimators", 100)
    mlflow.log_param("contamination", 0.05)
    mlflow.log_param("random_state", 42)
    
    isolation_forest = IsolationForest(
        n_estimators=100,
        contamination=0.05,
        random_state=42
    )
    isolation_forest.fit(x)
    
    x_full = data.drop(columns=ex_cols)
    x_full = x_full.reset_index(drop=True)
    
    decision_score = isolation_forest.decision_function(x_full)
    
    data['anomaly_score'] = decision_score
    
    healthy_score = decision_score[data['time'] <= 30]
    mean_healthy = healthy_score.mean()
    
    max_time = data.groupby('unit_number')['time'].transform('max')
    fail_score = decision_score[data['time'] > (max_time * 0.7)]
    mean_fail = fail_score.mean() if len(fail_score) > 0 else 0
    
    mlflow.log_metric("mean_healthy_score", mean_healthy)
    mlflow.log_metric("mean_failing_score", mean_fail)
    mlflow.log_metric("score_difference", mean_healthy - mean_fail)
    mlflow.log_metric("score_ratio", mean_fail / mean_healthy if mean_healthy != 0 else 0)
    
    models_dir = base_path / 'models'
    models_dir.mkdir(exist_ok=True)
    
    model_path = models_dir / 'anomaly_detector_if.joblib'
    joblib.dump(isolation_forest, model_path)
    mlflow.log_artifact(str(model_path))
   
    feature_names_path = models_dir / "feature_names.txt"
    with open(feature_names_path, "w") as f:
        f.write("\n".join(x.columns.tolist()))
    mlflow.log_artifact(str(feature_names_path))
    
    print(" MLflow run completed successfully!")

print(f"\n Results:")
print(f"   Mean healthy score: {mean_healthy:.4f}")
print(f"   Mean failing score: {mean_fail:.4f}")
print(f"   Score difference: {mean_healthy - mean_fail:.4f}")
```
